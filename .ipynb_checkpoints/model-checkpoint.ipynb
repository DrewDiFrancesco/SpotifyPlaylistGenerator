```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import numpy as np
import spotipy
from spotipy.oauth2 import SpotifyOAuth
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
# from creds import client_id, client_secret

featData = pd.read_csv("spotifySongFeatureData.csv")
FEATURE_KEYS = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo','duration','time_signature']
K_MAX = 200
```

```python
audio_feat = featData.drop(featData.columns[[0,1,2,3,17]],axis = 1)
```

```python
featData
```

```python
audio_feat
```

```python
scaler = StandardScaler()
audio_feat = scaler.fit_transform(audio_feat)
audio_feat_df = pd.DataFrame(audio_feat, columns = FEATURE_KEYS)
audio_feat_df
```

# Identify Number of Clusters:

```python
cost = list()
for i in range(2, K_MAX):
    KM = KMeans(n_clusters = i, max_iter = 500)
    KM.fit(audio_feat_df[FEATURE_KEYS])
    cost.append(KM.inertia_)

plt.plot(range(2, K_MAX), cost, color ='g', linewidth ='2')
plt.xlabel("Value of K")
plt.ylabel("Squared Error (Cost)")
plt.show()
```

# Using TSNE to reduce the dimensionality of audio vectors for cluster plotting:

```python
NUM_CLUSTERS = 10
TRACK_ADD_LIMIT = 100
COLORS = ['#e52165', '#0d1137', '#077b8a', '#5c3c92', '#1b6535',
          '#FF4040', '#E3CF57', '#FF9912', '#7FFF00', '#8B5742']
COLOR_MAP = {0:COLORS[0], 1:COLORS[1], 2:COLORS[2], 3:COLORS[3], 4:COLORS[4],
             5:COLORS[5], 6:COLORS[6], 7:COLORS[7], 8:COLORS[8], 9:COLORS[9]}

kmeans = KMeans(n_clusters=NUM_CLUSTERS, random_state=0)
featData['cluster'] = kmeans.fit_predict(featData[FEATURE_KEYS])
featData['color'] = featData.cluster.map(COLOR_MAP)

tsne = TSNE(n_components=2, random_state=0)
np.set_printoptions(suppress=True)
Y = tsne.fit_transform(featData[FEATURE_KEYS].values)
featData['x_coords'] = Y[:, 0]
featData['y_coords'] = Y[:, 1]
plt.scatter(featData.x_coords, featData.y_coords, c=featData.color, alpha = 0.6, s=10)
plt.show()
```
